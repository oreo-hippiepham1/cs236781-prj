## Data Structure
After taking a look at the data folder, as well as on the official website, it seems the data is structured as follows:  
- Actual images are stored in folders ```batch_1``` ... ```batch_15```  
- Information of images are stored in ```anotations.json```. The important infos are:  
    - *images*: Contains img_id, and file_path, used to retrieve the image  
    - *annotations*: Contains category_id for classification, segmentation data (polygon) and bboxes. Each entry is for 1 object annotation (so 1 image can have multiple entries). img_id is used as key to connect to the actual images

# Custom Dataset Class  
- Because of the rather complex structure of the data folder, we need to create a Dataset class that inherit

````python
from pycocotools import mask as mask_utils


class CustomDataset:
    def __init__(self, annotation_file, image_root):
        self.coco = COCO(os.path.join(annotation_file))
        self.image_root = image_root
        self.image_ids = list(self.coco.imgs.keys())

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        # retrieve image info
        image_id = self.image_ids[idx]
        image_info = self.coco.loadImgs(image_id)[0]
        image_path = os.path.join(self.image_root, image_info['file_name'])
        
        # load image
        image = cv2.imread(image_path)  

        # load annotations for current image
        annotation_ids = self.coco.getAnnIds(imgIds=image_id) # id
        annotations = self.coco.loadAnns(annotation_ids)

        # Extract bounding boxes, class labels, and segmentation masks
        boxes = [self._annotation_to_bbox(anno) for anno in annotations]
        labels = [anno['category_id'] for anno in annotations]
        masks = [anno['segmentation'] for anno in annotations]
        
        # append labels to end of bboxes (for Albumentation)
        # format is bbox_list = [[bbox1_x1, bbox1_x2, bbox1_y1, bbox1_y2, bbox1_label], [bbox2...],]
        bbox_extend = [box + [label] for box, label in zip(boxes, labels)]
        
        # Convert segmentation masks to binary masks
        binary_masks = []
        for mask in masks:
            binary_mask = self.decode_mask(mask, image_info)
            binary_masks.append(binary_mask)
            

        # Convert segmentation masks to binary masks
        mask_array = []
        for mask in masks:
            binary_mask = self.decode_mask(mask, image_info)
            mask_array.append(binary_mask)
        
        # Convert binary masks to a single mask tensor
        # mask_tensor = torch.stack([torch.from_numpy(mask.astype(float)) for mask in binary_masks])
        mask_tensor = torch.stack([self._convert_binary_mask(mask) for mask in mask_array])

        return image, boxes, bbox_extend, labels, mask_array, mask_tensor
        # return image, boxes, labels, mask_tensor
    
    
    def decode_mask(self, mask, image_info):
        if type(mask) == list:
            # If the mask is a list, use the COCO API to decode it
            rle = mask_utils.frPyObjects(mask, image_info['height'], image_info['width'])
            binary_mask = mask_utils.decode(rle)
        else:
            # If the mask is already a single binary mask, return it as is
            binary_mask = mask

        return binary_mask

    def _annotation_to_bbox(self, annotation):
        # Convert annotation to bounding box [x1, y1, x2, y2] format
        bbox = annotation['bbox']
        x1 = bbox[0]
        y1 = bbox[1]
        x2 = x1 + bbox[2]
        y2 = y1 + bbox[3]
        
        return [x1, y1, x2, y2]
    
    def _convert_binary_mask(self, mask):
        # Convert binary mask to tensor
        if len(mask.shape) == 2:
            # Add a channel dimension for grayscale masks
            mask = mask[:, :, None]
        elif len(mask.shape) == 3 and mask.shape[2] > 1:
            # Convert RGB masks to grayscale
            mask = mask.mean(axis=2, keepdims=True)
        return torch.from_numpy(mask.astype(float))
    
````

````python
import matplotlib.pyplot as plt
import numpy as np

def visualize_sample(image, boxes, labels, binary_masks):
    num_objects = len(boxes)

    fig, axes = plt.subplots(num_objects, 2, figsize=(10, 5 * num_objects))
    if num_objects > 1:
        axes = axes.flatten()

    for idx, (box, label, binary_mask) in enumerate(zip(boxes, labels, binary_masks)):
        # Display image with bounding boxes
        axes[idx * 2].imshow(image)
        axes[idx * 2].set_title('Image')
        axes[idx * 2].axis('off')
        axes[idx * 2].add_patch(
            plt.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1], fill=False, color='r')
        )
        axes[idx * 2].text(box[0], box[1] - 10, label, color='r', backgroundcolor='w')

        # Display binary mask
        axes[idx * 2 + 1].imshow(binary_mask, cmap='gray')
        axes[idx * 2 + 1].set_title('Segmentation Mask')
        axes[idx * 2 + 1].axis('off')

    plt.tight_layout()
    plt.show()

    
annotation_file = 'anno/annotations_main.json'
image_root = '/datasets/TACO-master/data'

# Instantiate the CustomDataset class
dataset = CustomDataset(annotation_file, image_root)

def viz_idx(idx):
    sample_image, sample_boxes, _, sample_labels, _,  sample_binary_masks = dataset[idx]
    visualize_sample(sample_image, sample_boxes, sample_labels, sample_binary_masks)
    
viz_idx(13)
````

- Now that it's working correctly, we'll use a slightly modified version of the class, which now only contains BBox and label information, suitable for object detection tasks (no more segmentation polygons)

````python
class ODDataset:
    def __init__(self, annotation_file, image_root, transform=None):
        self.coco = COCO(os.path.join(annotation_file))
        self.image_root = image_root
        self.image_ids = list(self.coco.imgs.keys())
        self.transform = transform

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        # Retrieve image info
        image_id = self.image_ids[idx]
        image_info = self.coco.loadImgs(image_id)[0]
        image_path = os.path.join(self.image_root, image_info['file_name'])
        
        # Load image
        image = cv2.imread(image_path)  

        # Load annotations for the current image
        annotation_ids = self.coco.getAnnIds(imgIds=image_id) # id
        annotations = self.coco.loadAnns(annotation_ids)

        # Extract bounding boxes and class labels
        bbox = [self._annotation_to_bbox_voc(anno) for anno in annotations]
        labels = [anno['category_id'] for anno in annotations]
        
        if self.transform is not None:
            bbox_extend = [box + [label] for box, label in zip(bbox, labels)]
            transformed = self.transform(image=image, bboxes=bbox_extend)
            image = transformed["image"]
            bbox_x = transformed["bboxes"] # format [x1, x2, y1, y2, label]
            bbox = [[box[0], box[1], box[2], box[3]] for box in bbox_x]
            labels = [box[4] for box in bbox_x]
            
            return image, bbox, labels

        return image, bbox, labels
    
    def _annotation_to_bbox_voc(self, annotation):
        # Convert annotation to bounding box [x1, y1, x2, y2] format
        bbox = annotation['bbox']
        x1 = bbox[0]
        y1 = bbox[1]
        x2 = x1 + bbox[2]
        y2 = y1 + bbox[3]

        return [x1, y1, x2, y2]
    
    def _annotation_to_bbox_coco(self, annotation):
        # Convert annotation to bounding box [x, y, width, height] format
        bbox = annotation['bbox']
        x = bbox[0]
        y = bbox[1]
        width = bbox[2]
        height = bbox[3]

        return [x, y, width, height]
````

### Data Augmentation  
- To have a more comprehensive training dataset, we will perform some data (image) augmentation.
- However, for an object detection task, we need to put the segmentation / detection data - the polygon and bounding box coordinates - through the same transformations that the actual images go through, to preserve their spatial accuracy.
- Thus, we will be using the library ```Albumentation``` (https://albumentations.ai/docs/getting_started/bounding_boxes_augmentation/) which is a powerful library that supports image augmentation for various tasks including but not limited to segmentation and object detection.

````python
import albumentations as A
import cv2
from albumentations.pytorch import ToTensorV2


mean = [.485, .456, .406]
std = [.229, .224, .225] # ImageNet normalization params

album_transform = A.Compose([
    A.Resize(1024, 1024),
    # A.RandomCrop(width=512, height=512), # this affects the data too much, as many bboxes get removed completely
    A.HorizontalFlip(p=0.6),
    A.RandomBrightnessContrast(p=0.3),
    A.Normalize(mean=mean, std=std),
    ToTensorV2(),
], bbox_params=A.BboxParams(format='pascal_voc'))
# The original format is "coco", however in the Dataset class we transformed into pascal voc format

# coco: x_top_left, y_top_left, height, width
# pascal_voc: x_top_left, y_top_left, x_bottom_right, y_bottom_right
# yolo: x_center, y_center, height_ratio, width_ratio
````

````python
import matplotlib.patches as patches
    
def plot_image_with_bboxes(image, bboxes, labels):
    """
    Plot an image with bounding boxes and labels.

    Args:
        image (numpy.ndarray): Input image as a NumPy array.
        bboxes (list): List of bounding boxes 
        labels (list): List of labels
    """
    # Create figure and axes
    fig, ax = plt.subplots(1)

    # Display the image
    try:
        ax.imshow(image)
    except TypeError:
        print('Not Showing correctly...')
        print(image.shape)
        image = image.permute(1, 2, 0).numpy()  # after transformations image become Tensor(c, h, w)
        ax.imshow(image)
        
    # Iterate over bounding boxes and labels
    for bbox, label in zip(bboxes, labels):
        x1, y1, x2, y2 = bbox[:4]

        # Calculate the width and height of the bounding box
        width = x2 - x1
        height = y2 - y1
        
        # Create a rectangle patch for the bounding box
        rect = patches.Rectangle((x1, y1), width, height, linewidth=1, edgecolor='r', facecolor='none')

        # Add the rectangle patch to the plot
        ax.add_patch(rect)

        # Add label text on top of the bounding box
        ax.text(x1, y1, str(label), color='r', verticalalignment='top',
                bbox={'color': 'white', 'alpha': 0.7, 'pad': 2})

    # Set axis limits to match image dimensions
    ax.set_xlim(0, image.shape[1])
    ax.set_ylim(image.shape[0], 0)

    # Remove the axis ticks
    ax.set_xticks([])
    ax.set_yticks([])

    # Show the plot
    plt.show()
````

````python
def test_augment(idx):
    dataset_no_transform = ODDataset(annotation_file, image_root, transform=None)
    img, bbox, labels = dataset_no_transform[idx]
    plot_image_with_bboxes(img, bbox, labels)
    
    dataset_transform = ODDataset(annotation_file, image_root, transform=album_transform)
    img, bbox, labels = dataset_transform[idx]
    plot_image_with_bboxes(img, bbox, labels)
     
annotation_file = 'anno/annotations_main.json'
image_root = '/datasets/TACO-master/data'

# Instantiate the CustomDataset class
test_augment(100)
````

Perfect! Seems like the Dataset class as well as data augmentation works correctly. We will stick to this for now, and if the model later on uses different formatting we will modify the class accordingly.

## Train Test Split, DataLoader

````python
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
````

````python
# Split the dataset indices into train, validation, and test sets
def train_val_test_split(dataset, valid_size=0.2, test_size=0.1, random_state=42):
    indices = list(range(len(dataset)))
    
    train_indices, test_indices = train_test_split(indices, test_size=test_size, random_state=random_state)
    train_indices, valid_indices = train_test_split(train_indices, test_size=valid_size/(1-test_size), random_state=random_state)
    
    train_dataset = torch.utils.data.Subset(dataset, train_indices)
    valid_dataset = torch.utils.data.Subset(dataset, valid_indices)
    test_dataset = torch.utils.data.Subset(dataset, test_indices)
    return train_dataset, valid_dataset, test_dataset
````

````python
# TRANSFORMATION
mean = [.485, .456, .406]
std = [.229, .224, .225] # ImageNet normalization params

album_transform = A.Compose([
    A.Resize(1024, 1024),
    # A.RandomCrop(width=512, height=512), # this affects the data too much, as many bboxes get removed completely
    A.HorizontalFlip(p=0.6),
    A.RandomBrightnessContrast(p=0.3),
    A.Normalize(mean=mean, std=std),
    ToTensorV2(),
], bbox_params=A.BboxParams(format='pascal_voc'))

# DATA ROOT
annotation_file = 'anno/annotations_main.json'
image_root = '/datasets/TACO-master/data'

# Create train, validation, and test datasets
ds_train, ds_valid, ds_test = train_val_test_split(ODDataset(annotation_file, image_root, album_transform))
````
